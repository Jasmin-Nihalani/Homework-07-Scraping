```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

#Had trouble with user agents. The website still shows 403. Used html pasted in slack.
```

```python
doc = BeautifulSoup(open('driver-license-locations.html', 'r').read())
```

```python
items = doc.find_all('tr')

rows = []

for item in items:
    row = {}
    columns = item.find_all('td')
    if columns[0].get_text(separator=" ", strip=True) == "Company":
        continue
    row["Company"] = columns[0].get_text(separator=" ", strip=True)
    row["Address"] = columns[1].get_text(separator=" ", strip=True)
    row["Telephone"] = columns[2].get_text(separator=" ", strip=True)
    row["Hours"] = columns[3].get_text(separator=" ", strip=True)
    
    rows.append(row)
```

```python
df = pd.json_normalize(rows)
df.head(10)
```

```python
df.to_csv("License.csv", index=False)
```
