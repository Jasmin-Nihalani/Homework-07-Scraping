```python
import pandas as pd
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States')
tables_new = tables[0]
```

```python
tables_new.drop(columns=['Portrait','Party[b][17]'],inplace=True)
```

```python
tables_new.rename(columns={"No.[a]":"No.","Term[16]":"Term","Party[b][17].1":"Party","Vice President[18]":"Vice_President"},inplace=True)
```

```python
tables_new.to_csv('Presidents_wiki.csv', index=False)
```
